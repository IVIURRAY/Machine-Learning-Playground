```python
# Import Librarys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
# Import the dataset
root = '/Users/Haydn/Documents/Code/Jupyter/Machine Learning A-Z/1.0 - Example Data/'
data_file = root + 'Part 3 - Classification/Logistic Regression/Social_Network_Ads.csv'
dataset = pd.read_csv(data_file)

'''
We want to find the correlation between this data set and their likelhood
of making the purchase.
'''
dataset
```

```python
x = dataset.iloc[:, [2, 3]].values # age and salary
y = dataset.iloc[:, 4].values # purchased?
```

```python
# Split dataset into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
print('X Training Size: %s\n' % len(x_train), x_train, '\nX Test Size: %s\n' % len(x_test), x_test, '\nY Training Size: %s\n' % len(y_train), y_train, '\nY Test Size: %s\n' % len(y_test), y_test)
```

```python
# Feature scaling
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
x_train
```

```python
# Build logistic regression model
classifier = LogisticRegression(random_state=0, solver='liblinear') # random state so we've same results as guy in video
classifier.fit(x_train, y_train)
```

```python
# Predict the test results
y_predictions = classifier.predict(x_test)
y_predictions
```

```python
# Lets check the results...evaluation time
matrix = confusion_matrix(y_test, y_predictions)
print('Right preditions:', matrix[0][0] + matrix[1][1])
print('Wrong preditions:', matrix[0][1] + matrix[1][0])
matrix
```

```python
# Graph it...Training Set
# Points on the graph are the users by age and salary.
# Red points is training set people who did not purchase
# Green point is training set people who did purchase
# The shaded regions are showing the classifier if they purchase or not.
# Stright line is the prediction boundary.
# If the line is stright then its a linear classifier.
# The wrong guesses from the classifier is because its linear
# its unable to account for the points as the line can't `bend`.
# It would need to be non-linear...check out later examples.
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(
    np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),
    np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01)
)
plt.contourf(
    x1, 
    x2, 
    classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
    alpha=0.75,
    cmap=ListedColormap(['red', 'green'])
)
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(
        x_set[y_set==j, 0],
        x_set[y_set==j, 1],
        c=ListedColormap(['red', 'green'])(i),
        label=j
    )
plt.title('Logistic Regression (Training Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
