```python
'''
Build multiple decision trees based on differnet subsets of the dataset.
Then when you add new points you ask all the trees what it puts it at.
Then you take the average of choices to classify.

Basically making lots of trees and then find the most common result.
'''
```

```python
# Import Librarys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
# Import the dataset
root = '/Users/Haydn/Documents/Code/Jupyter/Machine Learning A-Z/1.0 - Example Data/'
data_file = root + 'Part 3 - Classification/Logistic Regression/Social_Network_Ads.csv'
dataset = pd.read_csv(data_file)

'''
We want to find the correlation between this data set and their likelhood
of making the purchase.
'''
dataset
```

```python
x = dataset.iloc[:, [2, 3]].values # age and salary
y = dataset.iloc[:, 4].values # purchased?
```

```python
# Split dataset into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
print('X Training Size: %s\n' % len(x_train), x_train, '\nX Test Size: %s\n' % len(x_test), x_test, '\nY Training Size: %s\n' % len(y_train), y_train, '\nY Test Size: %s\n' % len(y_test), y_test)
```

```python
# Feature scaling
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
x_train
```

```python
# Build the model
no_of_trees = 10
classifier = RandomForestClassifier(n_estimators=no_of_trees, random_state=0)
classifier.fit(x_train, y_train)
```

```python
# Predict the test results
y_predictions = classifier.predict(x_test)
y_predictions
```

```python
# Lets check the results...evaluation time
matrix = confusion_matrix(y_test, y_predictions)
print('Right preditions:', matrix[0][0] + matrix[1][1])
print('Wrong preditions:', matrix[0][1] + matrix[1][0])
matrix
```

```python
# Graph it...Training Set
from matplotlib.colors import ListedColormap
plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k') # this makes the graph bigger...
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(
    np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),
    np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01)
)
plt.contourf(
    x1, 
    x2, 
    classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
    alpha=0.75,
    cmap=ListedColormap(['red', 'green'])
)
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(
        x_set[y_set==j, 0],
        x_set[y_set==j, 1],
        c=ListedColormap(['red', 'green'])(i),
        label=j
    )
plt.title('Logistic Regression (Training Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# Test Set Predictions
plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k') # this makes the graph bigger...
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(
    np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),
    np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01)
)
plt.contourf(
    x1, 
    x2, 
    classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
    alpha=0.76,
    cmap=ListedColormap(['red', 'green'])
)
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(
        x_set[y_set==j, 0],
        x_set[y_set==j, 1],
        c=ListedColormap(['red', 'green'])(i),
        label=j
    )
plt.title('Logistic Regression (Test Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
