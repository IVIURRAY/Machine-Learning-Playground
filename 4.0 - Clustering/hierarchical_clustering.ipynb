```python
'''
1. Make each point a datapoint
2. Take the two closest points and merge them into one cluster
3. Take the two closest clusters and merge them into one cluster
4. Repeat step 3 until only one cluster.
5. Finish

How to get the optimal clusters
1. Make a Dendrograms and find the larget distence between the bars and how many lines
   that bar goes though on the x axis is how many clusters you need.
'''
```

```python
# Import librarys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
'''
We have been hired to split the data about the customer in a mall to
split them into optimal categories based on their spending score in
this mall.

We will have to decide how many group to split these into.
'''
```

```python
# Import the dataset
root = '/Users/Haydn/Documents/Code/Jupyter/Machine Learning A-Z/1.0 - Example Data/'
data_file = root + 'Part 4 - Clustering/Mall_Customers.csv'
dataset = pd.read_csv(data_file)

# Spending score - the more or less that customer spends
dataset
```

```python
# Splitting features and outcomes
x = dataset.iloc[:, [3, 4]].values # anual income and spending score
# y = dataset.iloc[:, 3].values #
```

```python
# Make Dendrogram
dendrogram = sch.dendrogram(sch.linkage(x, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
```

```python
# Best clusters is 5 as this is the larget distance without crossing a horizontal line.
```

```python
# Fit hierarchical clustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(x)
y_hc
```

```python
# Visualise Clusters
# Graph it...
plt.scatter(x[y_hc==0, 0], x[y_hc==0, 1], s=100, c='red',     label='Careful')
plt.scatter(x[y_hc==1, 0], x[y_hc==1, 1], s=100, c='blue',    label='Standard')
plt.scatter(x[y_hc==2, 0], x[y_hc==2, 1], s=100, c='green',   label='Target')
plt.scatter(x[y_hc==3, 0], x[y_hc==3, 1], s=100, c='cyan',    label='Careless')
plt.scatter(x[y_hc==4, 0], x[y_hc==4, 1], s=100, c='magenta', label='Sensible')

plt.title('Clusters of clients')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
